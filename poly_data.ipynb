```python
import requests


GAMMA_API_BASE = "https://gamma-api.polymarket.com"

search = "bitcoin up or down"
status = "active" # "closed"
limit = 50
page = 1
resp = requests.get(f"{GAMMA_API_BASE}/public-search?events_status={status}&q={search}&limit_per_type={limit}&page={page}")
resp.raise_for_status()
chunk = resp.json()
print(chunk)
```
